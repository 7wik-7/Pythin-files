```
# Subplotting - Plot multiple graphs on same view

plt.figure(figsize=(10, 10))
plt.subplot(2,3,1)               # Syntax - [ROWS, COLS, INDEX_WHERE_TO_PLACE_THIS_PARTICULAR_GRAPH]
plt.plot(fmri.timepoint)

plt.subplot(2,3,3)
sns.histplot(data=fmri.signal)

plt.subplot(2,3,4)
fmri.region.value_counts().plot.bar()

plt.subplot(2,3,6)
fmri.event.value_counts().plot.pie()

plt.show()
```

```
# ML has 2 phase - 
# 1. Training (Training of a model)
# 2. Prediction  
```

```
#                     -------
# Input + Output --> [ML Algo] --> ML Model
#                     -------
# ML Model - can be understood as relationship between input Vs output that was harvested from trining data. 
# ML Model - is a big math equation
```
