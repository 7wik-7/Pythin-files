```python
### LOC SYNTAX ###
# df.loc[  condition for rows/row name ,   condition for columns /column name ]
```

```python
### SYNTAX FOR MULTIPLE CONDITION IN LOC ###
# df.loc[(condition1) & (condition2) & (condition3)]
```

```python
# When you want all conditions to be true please use &       (AND)
# When you want any one condition to be true please use |    (OR)
```

```python
df2.loc[(df2["class"]=="Normal") | (df2.pelvic_incidence > 65.0)]
```

```python
df2.loc[(df2.sacral_slope > 25) & (df2.sacral_slope < 35)]
```
